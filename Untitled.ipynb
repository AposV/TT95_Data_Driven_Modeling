```python
data_2019 = pd.read_csv('Prepared Datasets/Mar2019Aug2019rs1h.csv')
data_2022bm = pd.read_csv('Prepared Datasets/Mar2022Jun2022rs1h.csv')
data_2022am = pd.read_csv('Prepared Datasets/Aug2022Dec2022rs1h.csv')

lux_airport = pd.read_csv('../Data/Luxor Airport Weather Station/1943-2020_LuxorAirport_Hourly_RAW.csv')


data_2019.datetime = pd.to_datetime(data_2019.datetime)
data_2022bm.datetime = pd.to_datetime(data_2022bm.datetime)
data_2022am.datetime = pd.to_datetime(data_2022am.datetime)

data_2019.set_index('datetime', inplace=True)
data_2022bm.set_index('datetime', inplace=True)
data_2022am.set_index('datetime', inplace=True)
```

```python
ext1_app = pd.concat([data_2019['apperture_ext1'], data_2022bm['apperture_ext1'], data_2022am['apperture_ext1']])
ext2_app = pd.concat([data_2019['apperture_ext2'], data_2022am['apperture_ext2']])

ext1_d0 = data_2019['apperture_ext1'].iloc[0]
ext2_d0 = data_2019['apperture_ext2'].iloc[0]

data_2019['ext1_disp'] = data_2019['apperture_ext1'] - ext1_d0
data_2019['ext2_disp'] = data_2019['apperture_ext2'] - ext2_d0

data_2022bm['ext1_disp'] = data_2022bm['apperture_ext1'] - ext1_d0

data_2022am['ext1_disp'] = data_2022am['apperture_ext1'] - ext1_d0
data_2022am['ext2_disp'] = data_2022am['apperture_ext2'] - ext2_d0

```

```python
lux_airport_data = lux_airport.loc[:, ['DATE', 'TMP','DEW']]
lux_airport_data['TMP'] = lux_airport_data['TMP'].apply(lambda x: float(str(x).replace(',','.')[2:])/10)
lux_airport_data['DEW'] = lux_airport_data['DEW'].apply(lambda x: float(str(x).replace(',','.')[2:])/10)

lux_airport_data['DATE'] = pd.to_datetime(lux_airport_data['DATE'])
lux_airport_data = lux_airport_data.set_index('DATE')
```

```python
lux_airport_data = lux_airport_data[(lux_airport_data['TMP'] < 60) & (lux_airport_data['DEW'] < 60)]
lux_airport_data['TMP'].plot()
plt.title('Historical Temperature Measurements at Luxor Airport')
plt.ylabel('Temperature (degC)')
```

```python
# Calculate RH based on Dp according to: https://www.omnicalculator.com/physics/relative-humidity
lux_airport_data['RH'] = ((2.718**(17.625*lux_airport_data.DEW/(243.04+lux_airport_data.DEW)))/
            (2.718**(17.625*lux_airport_data.TMP/(243.04+lux_airport_data.TMP))))*100

lux_airport_data.loc['2019-04-15':'2019-08-16'].plot()
```
