```python
class Predictive_Model():
    
    def __init__(self, dataset, in_col_list, out_col_list):
        self.dataset = dataset
        self.model = keras.Sequential()
        
        self.set_inputs_outputs(in_col_list, out_col_list)
        

        self.model.add(layers.Dense(100, input_shape=(self.inputs.shape[1],)))
        self.model.add(layers.Dense(50))
        self.model.add(layers.Dense(1))
        self.model.compile(loss='mae', optimizer='adam')
        
        
    def set_inputs_outputs(self, in_col_list, out_col_list):
        self.inputs = self.dataset[in_col_list]
        self.output = self.dataset[out_col_list]
        
        self.x_scaler = StandardScaler().fit(self.inputs)
        self.y_scaler = StandardScaler().fit(self.output)
        
    def train_split(self, test_size=None, train_size=None, random_state=None, shuffle=False, stratify=None): 
        self.x_train, self.x_test, self.y_train, self.y_test = train_test_split(self.inputs,
                                                                                self.output,
                                                                                test_size=test_size,
                                                                                train_size=train_size,
                                                                                random_state=random_state,
                                                                                shuffle=shuffle,
                                                                                stratify=stratify)
        
    def fit(self, use_scaling=True):          
        if use_scaling:
            self.is_scaled = True
            self.history = self.model.fit(self.x_scaler.transform(self.x_train), self.y_scaler.transform(self.y_train))
            #self.is_scaled = True
        else:
            self.history = self.model.fit(self.x_train, self.y_train)
            self.is_scaled = False
            
        return self.history
                      
    def predict(self, X):
        if self.is_scaled:
            return self.y_scaler.inverse_transform(self.model.predict(X).reshape(-1,1))
        else:
            return self.model.predict(X)
        
    def predict_on_train_set(self):
        return self.model.predict(self.x_train)
        
    def predict_on_test_set(self):
        return self.model.predict(self.x_test)
```
