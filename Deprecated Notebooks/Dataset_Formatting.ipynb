```python
def swap_format(date, inplace = False):
    import_format = "%d-%m-%Y %H:%M:%S"
    export_format = "%Y-%m-%d %H:%M:%S"
    
    imp_date = datetime.strptime(date, import_format)
    exp_date = imp_date.strftime(export_format)
    if inplace:
        date = exp_date
    return exp_date

#wrong_format.Datetime = wrong_format.Datetime.apply(lambda x: swap_format(x))
```
